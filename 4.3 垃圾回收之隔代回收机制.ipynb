### gc 模块的使用

```python
import gc

gc.get_count()
# 获取当前自动执行垃圾回收的计数器，返回一个长度为3的列表

gc.get_threshold()
# 获取gc模块中自动执行垃圾回收的频率

gc.set_threshold(threshold0[,threshold1, threshold2])
# 设置自动执行垃圾回收的频率

gc.disable()
# python3默认开启gc机制，可以使用该方法手动关闭gc机制

gc.collect()
# 手动调用垃圾回收机制回收垃圾
```

```python
import sys 
import gc
import time


class TestObject(object):
    def __init__(self):
        print("当前的对象已经被创建， 占用的内存地址为: %s" %hex(id(self)))
    

    def __del__(self):
        print('当前对象马上被系统GC回收了')

gc.disable() # 关闭隔代回收机制， Python 3 中 是默认开启的
while True:
    a = TestObject()
    b = TestObject()
    a.pro =  b
    b.pro = a

    del a 
    del b 
    # print(sys.getrefcount(b))

    print(gc.get_threshold()) # 打印隔代回收的阈值

    print(gc.get_count()) # 打印垃圾回收的次数
    time.sleep(0.2)


```
